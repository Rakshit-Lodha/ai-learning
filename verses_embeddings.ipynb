```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
collection = chroma_client.create_collection(name = "gita_collection")
```

```python
verses = [ "You have the right to work only, but never to its fruits",
    "The mind can be your friend or your enemy",
    "Knowledge is the supreme purifier",
    "Be steadfast in yoga and perform your duty",
    "One who sees inaction in action is wise"]
```

```python
def get_embeddings(text):
    response = client.embeddings.create(
        input = text,
        model = "text-embedding-3-small"
    )

    embedding = response.data[0].embedding

    return embedding
```

```python
def cosine_similarity(vec_1, vec_2):
    dot = np.dot(vec_1, vec_2)
    a = np.linalg.norm(vec_1)
    b = np.linalg.norm(vec_2)
    return dot/(a * b)
```

```python
def user_input():

    user_query = input("What do you want to ask?")
    user_emb = get_embeddings(user_query)

    result = []
    
    for v in verses:
        verses_emb = get_embeddings(v)
        similarity = cosine_similarity(verses_emb, user_emb)
        result.append((similarity,v))

    sorted_result = sorted(result, reverse = True)

    return sorted_result[0][1]
```

```python
user_input()
```
