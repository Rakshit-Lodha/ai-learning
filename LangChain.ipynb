```python
llm_1 = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0)
```

```python
prompt = ChatPromptTemplate.from_messages(
    [("system", "You are an assistant which answers questions in one sentence or less"),
    ("user", "{question}")
    ])
```

```python
chain = prompt | llm_1
```

```python
response = chain.invoke({"question": "What is python?"})

response.content
```

```python
llm_2 = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an assistant which answers the questions in one sentence or less based on the {role}"),
        ("user", "{question}")
    ]
)

chain_1 = prompt | llm_2
```

```python
response_1 = chain_1.invoke({
    "role": "Answer in a pirate tone",
    "question": "What is life?"
})

response_1.content
```
