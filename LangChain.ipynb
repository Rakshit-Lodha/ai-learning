```python
llm_1 = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0)
```

```python
llm_2 = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an assistant which answers the questions in one sentence or less based on the {role}"),
        ("user", "{question}")
    ]
)

chain_1 = prompt | llm_2
```

```python
response_1 = chain_1.invoke({
    "role": "Answer in a pirate tone",
    "question": "What is life?"
})

response_1.content
```

```python
@tool

def multiply(a:int, b:int) -> int:
    """Multiply two numbers togehter"""
    return a * b

llm = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0
)

llm_tools = llm.bind_tools([multiply])

response = llm_tools.invoke("What is 5 times 3?")

response.tool_calls
```

```python
@tool

def multiply(a:int, b:int) -> int:
    """Multiply two numbers togehter"""
    return a * b

llm = ChatOpenAI(
    model = "gpt-3.5-turbo", temperature = 0
)

llm_tools = llm.bind_tools([multiply])

response = llm_tools.invoke("What is 5 times 3?")

response.tool_calls
```

```python
@tool

def multiply(a:int, b:int) -> int:
    """Multiply two numbers together"""
    return a*b

agent = create_agent(
    model = "gpt-3.5-turbo",
    tools = [multiply],
    system_prompt = "You are a mathematician, that just gives the answer"
)

response = agent.invoke(
    {"messages": [{"role": "user", "content": "What is 5 times 3?"}]}
)

response['messages'][-1].content
```
