```python
load_dotenv()
```

```python
from openai import OpenAI
```

```python
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
def get_embeddings(text):
    response = client.embeddings.create(
    input = text,
    model = "text-embedding-3-small")

    return response.data[0].embedding
```

```python
text_1 = "I hate pizza"
text_2 = "I love Pizza"
text_3 = "This sucks"
```

```python
print(text_1)
a = get_embeddings(text_1)
```

```python
print(text_2)
b = get_embeddings(text_2)
```

```python
print(text_3)
c = get_embeddings(text_3)
```

```python
f"""Cosine similarity tells us how similar two meanings are by measuring how aligned their vectors are, ignoring length.
"""
```

```python
def cosine_similarity(vec_1, vec_2):
    dot = np.dot(vec_1, vec_2)
    norm1 = np.linalg.norm(vec_1)
    norm2 = np.linalg.norm(vec_2)
    return dot/(norm1 * norm2)
```

```python
cosine_similarity(a,b)
```
