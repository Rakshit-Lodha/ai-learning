```python
from openai import OpenAI
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
reasons = ["Below average app compared to its peers. Too many ads. They keep promoting their genius through pop-ups, banners etc. I find it too cluttered. Other apps have a much cleaner and pleasing interface. Many features like downloading CG statement is available only for premium customers while other apps offer it for free. Overall not recommended.", "A massive letdown at the very beginning itself. This is by far the most frustrating and irritating app which I have ever experienced. I just opened the app to see what's new, as I had logged into my account after a long time on a new device, but the app has become so intrusive. It has put me into a KYC mode, wasting at least 10 minutes looping in form after form. I was so frustrated because I did not intend to sign up for any of the services beforehand", "I have been this app for 4 years and it used to be so good. It also had expense tracking. And was very simple to use. Now they have made it into an utter mess. I can't find my profit and loss reports (for tax filing). It does not allow you to check your credit score anymore. It has become more of a selling platform, wherein they charge you to even track external portfolios. This is insane. No other platforms do this. There are better platforms to switch to, like Groww, IndMoney, Kuvera, etc."]

company = 'ET Money'

tag_taxonomy = {
    "Product": ["Loss of features", "Feature Paywall", "Premium Features"],
    "UX": ["Cluttered Interface", "App Crash", "UI Issues"],
    "Competition": ["Competitor Comparison", "Feature Gap", "Market Position"],
    "Support": ["KYC Issue", "Support Issues", "Account Problems"],
    "Monetization": ["Too many Ads", "Premium Pricing", "Charges"]
}

for i in reasons:
    sentiment_analyser = client.chat.completions.create(
        model = "gpt-3.5-turbo",
        messages = [
            {"role": "system", "content": f"You will be getting a list of app reviews related to {company}, your job is to find out the sentiment of that review. Make sure that you are looking things in context with {company}. The JSON response should contain 2 things: 1. label: which needs to be either Positive/Negative or Neutral. 2. reason: you need to give me 2-3 tags which summarises the review. Use these tags from these categories: {tag_taxonomy}"},
            
            {"role": "user", "content": "I am getting crashes on login, support is unhelpful"},
            {"role": "assistant", "content": "{\"label\": \"negative\", \"reason\": [\"KYC Issue\", \"App Crash\"],\"type\": [\"Product\", \"Support\"]}}"},
    
            {"role": "user", "content": "The UI is interactive and fun"},
            {"role": "assistant", "content": "{\"label\": \"positive\", \"reason\": [\"UI Interactive\", \"Fun\"] \"product\": [\"UX\"]}"},
    
            {"role": "user", "content": "Zerodha and Groww are so much better than ET Money"},
            {"role": "assistant", "content": "{\"label\": \"negative\", \"reason\": [\"Competitors Better\", \"Comparative Loss\"], \"product\": [\"competition\", \"product\"]}}"},
    
            {"role": "user", "content": f"{i}"}
        ], temperature = 1, seed = 1
             )

    print(f"---------")
    print(f"Review: {i}")
    print("----------")
    print(sentiment_analyser.choices[0].message.content)
```
