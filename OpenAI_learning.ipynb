```python
temperature = [0, 0.7, 1.2, 1.3]

for i in temperature:
    response_better = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a storyteller, who gives creative one liner stories that has a start, end and climax"},
        {"role": "user", "content": "Write a bed time story about unicorns and startups"}
    ],
    temperature = i)

    print(f"With temperature: {i}")
    print(response_better.choices[0].message.content)
```

```python
print("Analysis")
print(f"With temp as 0: It gave probably the most straightfoward response \nWith temp as 0.7: It gave the output closest to the prompt, which is to write a story in 1 line \nWith temp as 1.2: The output was meh, nothing great and nothing bad \nwith temp as 1.3: This was the most creative")
```

```python
max_tokens = [30, 60, 120, 300, 500, 1000]

for m in max_tokens:
    response_max = client.chat.completions.create(
    model = "gpt-3.5-turbo",
        messages = [{"role": "system", "content": "You are a storyteller, who gives creative stories that has a start, end and climax"},
                    {"role": "user", "content": "Write a bed time story about unicorn and startups"}
        ], temperature = 0.7, max_tokens = m)
    print(f"Max Token Used: {m}")
    print(response_max.choices[0].message.content)
```

```python
response_multi = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages = [{"role": "user", "content": "Write a 3 line poem about rainbows and magic."}], temperature = 1, n = 5)

for i, choice in enumerate(response_multi.choices, 1):
    print("-------")
    print(f"Choice: {i}")
    print("------")
    print(choice.message.content)
```

```python
seed = [7, 42, 99]

for i in seed:
    response_seed_1 = client.chat.completions.create(
        model = "gpt-3.5-turbo",
        messages = [
            {"role": "user", "content": "Write a 1 liner bed time story about the universe"}
        ], temperature = 1, seed = i
    )
    print(f"-------")
    print(f"Seed: {i}")
    print(f"-------")
    print(response_seed_1.choices[0].message.content)
```
