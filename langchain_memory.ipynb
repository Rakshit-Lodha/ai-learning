```python
from langchain.agents import create_agent
```

```python
from langchain_core.tools import tool
```

```python
from langgraph.checkpoint.memory import InMemorySaver
```

```python
from dotenv import load_dotenv
```

```python
import os
```

```python
load_dotenv()
```

```python
from openai import OpenAI
```

```python
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
checkpointer = InMemorySaver()
```

```python
@tool

def compound_interest(amount: int, years: int) -> str:
    """Calculate the investment returns at 12% per annum returns compounded
    """
    print(f"Amount: {amount}, years: {years}")
    final_amount = amount*(1.12)**years
    result = f"₹{amount} invested for {years} @12% gives us {final_amount:.2f}"
    return result

agent = create_agent(
    model = "gpt-4o-mini",
    tools = [compound_interest],
    system_prompt = """You are a financial calculator. 
    When a user mentions an amount and years, immediately call the compound_interest tool.
    Do NOT ask for clarification if the numbers are already provided
    """,
    checkpointer = checkpointer
)

config = {"configurable": {"thread_id": "user_1"}}

response = agent.invoke(
    {"messages": [{"role": "user", "content": "Calculate for 100000 rupees and 15 years"}]},
    config = config
)

response['messages'][-1].content
```

```python
response_2 = agent.invoke(
    {"messages": [{"role": "user", "content": "What would be the value in 1 year?"}]},
    config = config
)

response_2['messages'][-1].content
```
